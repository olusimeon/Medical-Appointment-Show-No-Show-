```python
# importing libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# renaming mis-splet column name
df=df.rename(columns={"PatientId":"Patient_id","AppointmentID":"Appointment_id","Gender":"Gender","ScheduledDay":"Scheduled_day",
                  "AppointmentDay":"Appointment_day","Age":"Age", "Neighbourhood":"Neighbourhood","Scholarship":"Scholarship", "Hipertension":"Hypertension",
                  "Diabetes":"Diabetes","Alcoholism":"Alcoholism","Handcap":"Handicap","SMS_received":"SMS_received","No-show":"No_show"})
df.head()
```

```python
# dropping irrelevant columns 
df=df.drop(columns=["Patient_id"])
```

```python
# cleanig the data set to eliminiate negative values
df.query("Age >=0", inplace= True)
```

```python
df["Age"].describe()
```

```python
# changing the scheduledDay to real date format 
df["Scheduled_day"]=pd.to_datetime(df.Scheduled_day)
df["Scheduled_day"]
```

```python
# changing the AppointmentDay to real date format 
df["Appointment_day"]= pd.to_datetime(df.Appointment_day)
df["Appointment_day"]
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
# binning our continuous variable to groups
binwidth= int((max(df["Age"]) - min(df["Age"]))/5)
bins=[0,18,37,55,115]
group=["0-17","18-36","37-54","55+"]
df["Age_binned"]=pd.cut(df["Age"],bins,labels=group,include_lowest=True)
```

```python
df.Age_binned.value_counts()
```

```python
#determining days between appointment day and schedule day
df["Days_diff"]=(df["Appointment_day"] - df["Scheduled_day"]).astype("timedelta64[D]")
```

```python
df["Days_diff"].describe()
```

```python
df.Days_diff.value_counts()
```

```python
df[df.Days_diff < 0]["Days_diff"].unique()
```

```python
df.query("Days_diff >= -1", inplace = True)
```

```python
df["Days_diff"].describe()
```

```python
df.Days_diff.value_counts()
```

```python
df["Days_diff"].describe()
```

```python
#time difference is not considered since the date of schedule and appointment date are the same but time defers maybe due to inacurate entering thus the negative values are changed to zero
df.Days_diff.replace(-1.0, 0, inplace=True)
```

```python
df["Days_diff"].describe()
```

```python
# binning days difference using the percentile
categories= pd.Series(["short:0-2", "week: 3-13","long: 14 +"])
```

```python
bins=[0,3,14,180]
df["Days_diff_binned"]= pd.cut(df["Days_diff"],bins,labels=categories)
```

```python
df.head(-10)
```

```python
# fill missing values using date difference to assign bin area
df.Days_diff_binned.fillna("short:0-2", inplace=True)
```

```python
df["Days_diff_binned"].value_counts()
```

```python
df.head(10)
```

```python
# converting appointment day of the week to actual day for proper analysis
df["Appointment_DOW"]=df.Appointment_day.dt.dayofweek
df["Scheduled_DOW"]=df.Scheduled_day.dt.dayofweek
```

```python
# assigning days
days=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]

def dayNameFromWeekday(weekday):
    days_=[]
    for day in weekday:
        days_.append(days[day])
    return days_
    
```

```python
df["Appointment_DOW"] =dayNameFromWeekday(df.Appointment_DOW)
df["Scheduled_DOW"] =dayNameFromWeekday(df.Scheduled_DOW)
```

```python
categories=days
df["Appointment_DOW"]= pd.Categorical(df.Appointment_DOW, categories=days,ordered=True)
df["Scheduled_DOW"]= pd.Categorical(df.Scheduled_DOW, categories,ordered=True)
```

```python
# assigning X and y variables for prediction
X=df[["Gender","Diabetes","Hypertension","Scholarship","SMS_received","Handicap","Age","Days_diff"]]
y=df["No_show"]
```

```python
X.head()
```

```python
y
```

```python
X_train=pd.get_dummies(X)
X_train,X_test,y_train,y_test=train_test_split(X_train,y,test_size=0.25)
```

```python
clf = LogisticRegression()
clf.fit(X_train, y_train)
y_pred_log_reg = clf.predict(X_test)
acc_log_reg = round( clf.score(X_train, y_train) * 100)
print (str(acc_log_reg) + ' percent')
```

```python
logreg= LogisticRegression()
logreg.fit(X_train,y_train)
y_pred=logreg.predict(X_test)

print("Results: ")
print("Accuracy", metrics.accuracy_score(y_test,y_pred))

confusion= metrics.confusion_matrix(y_test,y_pred)

TP=confusion[1,1]
TN=confusion[0,0]
FP=confusion[0,1]
FN =confusion[1,0]

print("specificity:", TN / float (TN +FP))
print("False positive rate: ", FP / float(TN + FP))

#print("Recall: ", metrics.recall_score(y_test,y_pred))

```
